```python
import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
#Suppose if i have extra header 
df = pd.read_csv('weather_data.csv')
df
```

```python
#use skip rows
df = pd.read_csv('weather_data.csv', skiprows=1)
df
```

```python
#we can also do the same thing using header

df = pd.read_csv('weather_data.csv', header=1) #because index starts from 0 so, header =1 means it's 2nd row in csv file
df
```

```python
#suppose we don't have header then
df = pd.read_csv('weather_data.csv', header=None)
df
```

```python
df = pd.read_csv('weather_data.csv', header=None, names=['day', 'temperature', 'windspeed', 'event'])
df
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
#argument na_values

df = pd.read_csv('weather_data.csv', na_values=['n.a', 'not available']) #na_values when ever 'n.a' or 'not available' is encountered it will replace with NaN
df
```

```python
#another way of supplying na_vaules using dictionary

df = pd.read_csv('weather_data.csv', na_values={'temperature': ['n.a'], 'windspeed': ['not available']})
df
```

```python
#writing dataframe into new file
df.to_csv('new.csv') #default index will also write into new file
```

```python
#if you don't want index 
df.to_csv('new.csv', index=False)
```

```python
#if you don't want to write all columns (skip columns)
df.columns
```

```python
df.to_csv('new.csv', columns=['day', 'temperature'], index=False)

# if you want to skip header you can use argument header=False
```
