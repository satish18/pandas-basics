# Handling Missing Data 

fillna

dropna

interpolate

Topics: 

1. fillna to fill missing values using different ways

2. interpolate to make a guess on missing values using interpolation

3. dropna to drop rows with missing values

```python
import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
type(df['day'][0])
```

```python
#convert string to date while reading as follows
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df
```

```python
type(df['day'][0])
```

```python
#set index as day

df.set_index('day', inplace=True)
df
```

```python
#want to replace all NaN with some value let's say 0

new_df = df.fillna(0)
new_df
```

```python
#this is wrong event should be either snow or Sunny or cloudy for that you can specify according to column as follows

new_df = df.fillna({'temperature': 0, 'windspeed': 0, 'event': 'no event'})
new_df
```

```python
#some times while calculating mean or something like that 0 would be probably wrong guess for that we can use carry forward previous values
#or we can use back bfill
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

#if you want more properties on fillna 

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.fillna.html

```python
#let's see what axis argument will do on fillna
new_df = df.fillna(method='bfill', axis="columns")
new_df #it will copy horizontal back fill
```

```python
#limit argument
new_df = df.fillna(method='ffill', limit=1) #limit for forward fill only for one row
new_df
```

```python
#if want some better guess of missing values you can use interpolate

#it uses linear interpolation

new_df = df.interpolate()
new_df
```

#For more properties follow the link

http://pandas.pydata.org/pandas-docs/version/0.16.2/generated/pandas.DataFrame.interpolate.html

```python
new_df = df.interpolate(method='time')
new_df
```

```python
#if you want to drop all rows which have NaN
new_df = df.dropna() #it will drop rows which has atleast one NaN is present
new_df
```

```python
#if you want to drop rows where all entries of row are NaN

new_df = df.dropna(how='all')
new_df
```

```python
#we can also do the above thing using following cmd
new_df = df.dropna(thresh=1)
new_df
```
